```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
from keras import Input
from keras.regularizers import l2
from keras.layers import Dropout
```

```python
df = pd.read_csv("C:/Users/Dominic/Documents/UNSW/2022H5 Data mining and Machine Learning/week_3/pima/pima.csv", header=None)
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1:]
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42)
```

```python
X_train
```

```python
model = Sequential()
model.add(Dense(9, input_dim=X_train.shape[1], kernel_regularizer=l2(0.001)))
model.add(Dense(9, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='SGD', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, validation_data=(x_test, y_test), epochs=500, batch_size=10)
```

```python
_, acc_train = model.evaluate(X_train, y_train)
_, acc_test = model.evaluate(x_test, y_test)
```
