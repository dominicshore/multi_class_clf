```python
from keras.layers import Dense
from keras.models import Sequential
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from matplotlib import pyplot

import pandas as pd
```

```python
df = pd.read_csv("~/Documents/UNSW/2022H5 Data mining and Machine Learning/week_3/fraud/creditcard.csv")
df
```

```python
target_column = ['Class']
predictors = list(set(list(df.columns)) - set(target_column))
X = df[predictors].values
y = df[target_column].values
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=41)
```

```python
model = Sequential()
model.add(Dense(units = 60, input_dim = 30, activation='relu'))
model.add(Dense(units = 60, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
mod_fit = model.fit(X_train, y_train, validation_data=(x_test, y_test), epochs=10, verbose=0)
```

```python
# Evaluate the model
_, train_acc = model.evaluate(X_train, y_train, verbose=0)
_, test_acc = model.evaluate(x_test, y_test, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
pyplot.plot(mod_fit.history['accuracy'], label='train')
pyplot.plot(mod_fit.history['val_accuracy'], label='test')
pyplot.legend()
```
