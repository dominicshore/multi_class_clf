```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import export_text
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/Dominic/Documents/UNSW/2022H5 Data mining and Machine Learning/week_3/tennis/tennis.csv")
df.head()
```

```python
df.iloc[:, 1:-1]
```

```python
featureframe = df.iloc[:, 1:-1]

targetframe = df.iloc[:, -1:]
```

```python
dummy_cols = [ 'outlook', 'temp', 'humidity', 'wind'] 
```

```python
ff = pd.get_dummies(featureframe, columns=dummy_cols, drop_first=True) 
feature_names = ff.columns
features = pd.DataFrame(ff).to_numpy( )
```

```python
tf = pd.get_dummies(targetframe, columns=['play'] , drop_first=True) 
target_name = tf.columns
target = pd.DataFrame(tf).to_numpy( )
target = np.ravel(target)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state=42)
```

```python
for i in [X_train, X_test, y_train, y_test]:
    print(i.shape)
```

```python
tree_model = DecisionTreeClassifier(max_features="sqrt", 
                                    random_state=42, 
                                    min_samples_leaf=1, 
                                    max_depth=20)
```

```python
tree_model.fit(X_train, y_train)
```

```python
r = export_text(tree_model, show_weights=True, feature_names=list(feature_names))
print(r)
```

```python
target_predicted = tree_model.predict(X_test)
```

```python
con_mat = confusion_matrix(y_test, target_predicted)
```
